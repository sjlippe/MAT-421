# 17.1 - Interpolation Problem Statement

The problem statement for interpolation states that if we have a data set consisting of [x_i , y_i], our goal is to find something called the estimation function, y_hat(x) such that y_hat(x_i) = y_i. This describes the process of interpolation where given a new x value, we can interpolate the output function value of the new x value. The graph belows outlines an example where we are predicting the y value at an x value of 4.


```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [1,2,3,5,6]
y = [2,4,6,3,7]
plt.figure(figsize = (12,7))
plt.plot(x,y,'bo', 4,4.5,'ro')
plt.title('Data Points for Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 17.2 - Linear Interpolation

Through linear interpolation, we can assume that the value we want to interpolate for lies on the same line to the left of the next highest value and to the right of the previous lower value. The linear interpolation can then be calculated using the function:

y(x) = y_i + ( ( (y_i + 1 - y_1)*(x - x_i) ) / (x_i + 1 - x_i) )

```python
# Lets say we wanted to find the interpolate at x = 4.5 based on the data set below

from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [0, 1, 2, 4, 6]
y = [1, 3, 2 ,5, 7]

f = interp1d(x, y)
y_hat = f(4.5)
print(y_hat)
```

```python
# Now lets further demonstrate this through a graph

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(4.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 4.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# We can see that the corresponding y value at an x value of 4.5 is exactly 5.5.
```

# 17.3 - Cubic Spline Interpolation

Cubic spline interpolation is much different than linear interpolation in which the interpolating function is actually a set of piecewise cubic functions. We assume that the points (x_i, y_i) and (x_i+1 , y_i+1) are joined by a cubic polynomial:
S_i(x) = a_i*x^3 + b_i*x^2 + c_i*x + d_i

In order to achieve cubic spline interpolation, each coefficient must be determined. We have a total of 4(n-1) unknowns which means we need 4(n-1) equations to find the coefficients.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [0, 1, 2, 3]
y = [1, 3, 3.2, 1.9]

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 3, 100)
y_new = f(x_new)
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
